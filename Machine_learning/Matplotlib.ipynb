### Two optinos for subplots 

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10, 5))
#ploting to each axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot to different axisusing index
ax[0, 0].plot(x,x*2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which polt to use ? (pyplot Vs matplotlib OO method)
* when ploting something quickly it's ok to use pyplot method
* When ploting something more advance use maplotlib.

```python
over_50 = heart_disease[heart_disease.age > 50]
len(over_50)
```

```python
# pyplot method
over_50.plot(kind="scatter",x='age',y='chol',c='target');
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind = 'scatter',x='age',y='chol',c='target',ax=ax);
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))
#plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

#Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
        xlabel="Age",ylabel="Cholestrol")
#Ading a legend
ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50["chol"].mean(),linestyle='--');


```

```python
# Subplot of chol, age, thalac
fig, (ax0, ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10, 10))

#Adding data to ax0
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])

#Customize ax0
ax0.set(title="Heart Disease and cholestrol Levels",
        xlabel ='Age',
        ylabel='Cholestrol')
#Add legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");
#Add meanline
ax0.axhline(over_50["chol"].mean(),linestyle='--');

#Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

#customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
xlabel="Age",
ylabel="Max Heart Rate");
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
#Add Mean line
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");
#add title to figure
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
```

```python
plt.style.use('seaborn-whitegrid')
#OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))
#plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap="winter") #this changes the colousr scheme

#Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
        xlabel="Age",ylabel="Cholestrol")
#Ading a legend
ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
#customizing the x and y limits
# Subplot of chol, age, thalac
fig, (ax0, ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10, 10))

#Adding data to ax0
scatter = ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap="winter")

#Customize ax0
ax0.set(title="Heart Disease and cholestrol Levels",
        xlabel ='Age',
        ylabel='Cholestrol')
#change x axis limits
ax0.set_xlim([50, 80])
#Add legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");
#Add meanline
ax0.axhline(over_50["chol"].mean(),linestyle='--');

#Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")

#customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
xlabel="Age",
ylabel="Max Heart Rate");
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
#Add Mean line
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");
#add title to figure
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold");
```
